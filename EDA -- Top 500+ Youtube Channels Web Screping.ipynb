```python
Rank = []
ChannelName = []
Category = []
Subscribers = []
Country = []
Views = []
Likes = []
Comments = []

for i in range(1,21):
    urls = "https://hypeauditor.com/top-youtube/?p="+str(i)
    print(urls)
    page = requests.get(urls)
    soup = BeautifulSoup(page.text)
 
    # rank
    a = soup.find_all("div",class_ = "row-cell rank")
    for i in a:
    
        Rank.append(i.text)

    # Channelnames
    b = soup.find_all("div",class_ = "contributor__name-content")

    for i in b:
        a = i.text
        if a == '':
            ChannelName.append(np.nan)
        else:
            ChannelName.append(i.text)
    
    # category
    c = soup.find_all("div",class_ = "contributor__title")
    for i in c:
        Category.append(i.text)
    
    # subscribers
    d = soup.find_all("div",class_ = "row-cell subscribers")
    for i in d:
    
        Subscribers.append(i.text)
        
    # country    
    e = soup.find_all("div",class_ = "row-cell audience")
    for i in e:
        a = i.text
        
        if a == '':
            Country.append(np.nan)
        else:
            Country.append(a)
            
    # views        
    f = soup.find_all("div",class_ = "row-cell avg-views")
    for i in f:
       
         Views.append(i.text)
        
    # likes    
    g = soup.find_all("div",class_ = "row-cell avg-likes")
    for i in g:
        
        Likes.append(i.text)
        
    # comments    
    h = soup.find_all("div",class_ = "row-cell avg-comments")
    for i in h:
        
        Comments.append(i.text)
```

```python
len(Rank),len(ChannelName),len(Category),len(Subscribers),len(Country),len(Views),len(Likes),len(Comments)
```

```python
df_youtube = pd.DataFrame({"Rank":Rank,
    "ChannelName":ChannelName,
    "Category":Category,
    "Subscribers":Subscribers,
    "Country":Country,
    "Views":Views,
    "Likes":Likes,
    "Comments":Comments})
df_youtube
```

```python
df = pd.read_csv('youtube_webscraping.csv')
df
```

```python
# drop Unnamed columns
df = df.drop("Unnamed: 0",axis=1)
```

```python
df
```

```python
# removeing Special Symbols
df['Category'] = df['Category'].replace({r'[^\w\s]': ''}, regex=True)
df['Rank'] = df['Rank'].replace({r'[^\w\s]': ''}, regex=True)
df
```

```python
# converting K,M to 0000
df["Subscribers"] = df["Subscribers"].replace({"M": "*1e6"}, regex = True).map(pd.eval).astype(int)
df["Views"] = df["Views"].replace({"K": "*1e3","M": "*1e6"}, regex = True).map(pd.eval).astype(int)
df["Likes"] = df["Likes"].replace({"K": "*1e3","M": "*1e6"}, regex = True).map(pd.eval).astype(int)
df["Comments"] = df["Comments"].replace({"K": "*1e3"}, regex = True).map(pd.eval).astype(int)
```

```python
# Sample data (you can use your own dataset)
data = {
    'ChannelName': ['TSeries', 'MrBeast', 'SET India', 'PewDiePie', 'Zee Music Company', 'VYRLOriginals', 'Mussoumano', 'Myke Towers', 'Red Chillies Entertainment', 'Ankita Bhawsar'],
    'Category': ['TSeries', 'MrBeast', 'SET India', 'PewDiePie', 'Zee Music Company', 'VYRLOriginals', 'Mussoumano', 'Myke Towers', 'Red Chillies Entertainment', 'Ankita Bhawsar'],
    'Country': ['India', 'United States', 'India', 'United States', 'India', 'India', 'Brazil', 'Colombia', 'India', 'India']
}

df = pd.DataFrame(data)

# Set up a multi-plot grid
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plotting based on Category
sns.countplot(data=df, x='Category', palette='Set3', ax=ax1)
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45)
ax1.set_title('ChannelName Count by Category')
ax1.set_xlabel('Category')
ax1.set_ylabel('Count')

# Plotting based on Country
sns.countplot(data=df, x='Country', palette='Set2', ax=ax2)
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45)
ax2.set_title('ChannelName Count by Country')
ax2.set_xlabel('Country')
ax2.set_ylabel('Count')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

* Top YouTube channels dominated by Indian channels, with TSeries leading the way.

```python
# Exclude rows with missing values in 'Country' or 'Subscribers' columns
df_filtered = df.dropna(subset=['Country', 'Subscribers'])

# Convert the 'Subscribers' column to string
df_filtered['Subscribers'] = df_filtered['Subscribers'].astype(str)

# Remove 'M' from 'Subscribers' column and convert to float
df_filtered['Subscribers'] = df_filtered['Subscribers'].str.replace('M', '').astype(float)

# Group by 'Country' and sum the 'Subscribers' column
country_subscribers = df_filtered.groupby('Country')['Subscribers'].sum().reset_index()

# Sort the DataFrame by 'Subscribers' column in descending order
country_subscribers_sorted = country_subscribers.sort_values('Subscribers', ascending=False).head(10)

# Create the bar chart using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Subscribers', y='Country', data=country_subscribers_sorted)

# Set the plot title and axis labels
plt.title('Top 10 Countries by Subscriber Count')
plt.xlabel('Subscriber Count (in millions)')
plt.ylabel('Country')

# Display the plot
plt.show()
```

* Indian channels dominate the top 10 countries by YouTube subscriber count.

```python

# Box plots
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Subscribers', 'Views', 'Likes', 'Comments']])
plt.xlabel('Columns')
plt.ylabel('Values')
plt.title('Box Plots of Subscribers, Views, Likes, Comments')
plt.show()
```

```python
# Select the desired columns
subset = df[['Subscribers', 'Views', 'Likes', 'Comments']]

# Calculate the correlation matrix
correlation_matrix = subset.corr()

# Create a heatmap with the VIBGYOR color scheme
cmap = sns.color_palette("hsv", 7)
sns.heatmap(correlation_matrix, cmap=cmap, annot=True)

# Add title and axis labels
plt.title('Correlation Heatmap')
plt.xlabel('Variables')
plt.ylabel('Variables')

# Display the plot
plt.show()
```

* Correlation heatmap shows strong positive correlation between views, likes, and comments for YouTube channels.
